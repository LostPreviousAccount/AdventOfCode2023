```julia
function travel(step, key, instructions, dict)
  # println(step,'-', key,'-', instructions)
  if key == "ZZZ"
    step
  else
    travel(step+1, get(dict, key, 0)[instructions[1]] ,circshift(instructions,-1), dict)
  end
end

function solution1(input, startNode)
  lines = split(input, "\n")
  instructions = [l=='R' ? 2 : 1 for l in String(lines[1])]
  networkTemp = [split(lines[line], " = ") for line = 3:length(lines)]
  network = Dict(map(t -> (t[1], split(t[2][2:end-1],", ")),networkTemp))
  travel(0, startNode, instructions, network)
end
```

```julia
function travel2(step, key, instructions, dict, endWithZ, force)
  while force || !(key in endWithZ)
    # println(step,'-', keys,'-', instructions[1])
    step += 1
    key = get(dict, key, 0)[instructions[1]]
    instructions = circshift(instructions,-1)
    force = false
  end
  #(step, key, instructions)
  step
end

function nbLoops(key, instructions, network, endWithZ)
  firstTravel = travel2(0, key, instructions, network, endWithZ, false)
  println(firstTravel[1], firstTravel[2])
  #println(length(instructions) / firstTravel[1], firstTravel[2])
  println(instructions == firstTravel[3])
  theLoop = travel2(0, firstTravel[2], instructions, network, endWithZ, true)
  println(theLoop[1], theLoop[2])
  #println(length(instructions) / theLoop[1], theLoop[2])
  println(instructions == theLoop[3])
end

function solution2(input)
  lines = split(input, "\n")
  instructions = [l=='R' ? 2 : 1 for l in String(lines[1])]
  networkTemp = [split(lines[line], " = ") for line = 3:length(lines)]
  network = Dict(map(t -> (t[1], split(t[2][2:end-1],", ")),networkTemp))
  endWithA = filter(n -> n[3]=='A', keys(network)) |> collect
  endWithZ = filter(n -> n[3]=='Z', keys(network)) |> collect
  println(endWithA)
  println(endWithZ)
  println(length(instructions))  
  println(map(start -> travel2(0, start, instructions, network, endWithZ, false),endWithA))
end
```
